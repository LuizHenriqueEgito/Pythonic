# Dicionários (HashMap)

`dicionarios (dict)` é uma estrutura de dados que possui uma **chave** e um **valor** associado. As **chaves** precisam ser unicas não podem repetir, já os valores aceitam qualquer tipo de dado inclusive outro `dict`. O simbolo para criação de um dicionário é dado por `{}`, para criar um dicionario fazemos:

```python
dict_ = {}  # dicionario vazio
dict_['chave_1'] = 'valor_1'  # cria uma chave e um valor
# para acessar o valor fazemos
dict_['chave_1']
```

```python
# outra forma, aqui nos já criamos o dicionario e adicionamos valores a ele
dict_ = {
    'chave_1': 'valor_1',
    'chave_1': 'valor_1',
    'dicionario_interno': {
        'chave_interna_1': 'valor_interna_1',
        'chave_interna_2': 'valor_interna_2',
    }
}
dict_['chave_1']
```

```python
print(dict_['dicionario_interno'])
print('\nPegando um item interno (que está dentro do nosso dicionario_interno)')
print(dict_['dicionario_interno']['chave_interna_1'])
```

Como no python tudo é um `objeto` com dicionários isso não é diferente, e por serem `objetos` eles possuem `methods (métodos)` e `attributes (atributos)`.

```python
print(dir(dict_))
```
