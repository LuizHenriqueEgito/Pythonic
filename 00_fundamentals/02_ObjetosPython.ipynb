# Objetos

Em `Python` tudo é um **objeto**

```python
string = 'a'
print(type(string))

number = 1.0
print(type(number))

def f(): pass
print(type(f))
```

Se os *tipos* delas são **class** logo elas são `objetos`, e como `objetos` elas possuem:
- Métodos: "Verbos | Ações" que o objeto pode fazer;
- Atributos: "Adjetivos" do objeto, não executam ações eles são apenas uma informação.

Com o **dir(objeto)** no python podemos ver todos os métodos e atributos que esse objeto possui.

```python
print(dir(number))
print('\nFazendo a mesma coisa mas de uma forma diferente:')
print(number.__dir__())
```

Agora perceba como é o funcionamento do `Python`:

```python
print(number + 2)
print(number.__add__(2))
```

Isso porque `__add__` é o método magico que permite o python entender o que é o operador `+` entre dois numeros.

```python
number.__str__()
```
