# __dunder__ (Méetodos Mágicos)
Os métodos mágicos *(dunder)* são métodos especiais, que permitem sobrecarga de operadores (isso é bem interessante, permite que você faça:
```python
'Olá ' + 'mundo!'
```
E personalize seus objetos.
## Sobrecarga de Operadores
Permite que `+`, `-`, `*`, `/` e outros funcionem nos seus próprios objetos customizados.

````python
class Vetor:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # sobrecarga de +
    def __add__(self, other):
        return Vetor(self.x + other.x, self.y + other.y)

    def __repr__(self):
        return f"Vetor({self.x}, {self.y})"
````

````python
# Aqui o + chama __add__
v1 = Vetor(1, 2)
v2 = Vetor(3, 4)

print(v1 + v2)
````

````python
v1.__add__(v2)
````

# Métodos Mágicos (Dunder Methods) em Python

## 1. Inicialização e Construção
- `__new__(cls, ...)`  
  Responsável por **criar a instância** (baixo nível, raro de usar).

- `__init__(self, ...)`  
  Inicializa o objeto após a criação (mais comum).

- `__del__(self)`  
  Destruidor do objeto (evite usar — comportamento não é confiável).

---

## 2. Métodos Numéricos
- `__trunc__(self)`  
  Define o comportamento de `math.trunc()`.

- `__round__(self, n)`  
  Define o comportamento de `round()`.

- `__abs__(self)`  
  Define o comportamento de `abs()`.

- `__neg__(self)`  
  Define o comportamento do operador `-obj`.

- `__pos__(self)`  
  Define o comportamento do operador `+obj`.

---

## 3. Operadores Aritméticos
- `__add__(self, other)` → `+`  
- `__sub__(self, other)` → `-`  
- `__mul__(self, other)` → `*`  
- `__truediv__(self, other)` → `/`  
- `__mod__(self, other)` → `%`  

---

## 4. Métodos de String e Representação
- `__str__(self)`  
  Representação legível para humanos (`str()`).

- `__repr__(self)`  
  Representação oficial (debug, idealmente reproduzível).

- `__hash__(self)`  
  Retorna um inteiro usado em estruturas como `dict` e `set`.

- `__bool__(self)`  
  Define o valor booleano (`True` ou `False`).

- `__dir__(self)`  
  Lista atributos disponíveis (`dir()`).

---

## 5. Métodos de Comparação
- `__eq__(self, other)` → `==`  
- `__ne__(self, other)` → `!=`  
- `__lt__(self, other)` → `<`  
- `__gt__(self, other)` → `>`  
- `__ge__(self, other)` → `>=`

---

## 6. Métodos de Coleções e Iteração
- `__len__(self)`  
  Define o comportamento de `len(obj)`.

- `__getitem__(self, key)`  
  Permite acessar elementos com `obj[key]`.

- `__setitem__(self, key, value)` *(extra útil)*  
  Permite atribuição `obj[key] = value`.

- `__delitem__(self, key)` *(extra útil)*  
  Permite deletar com `del obj[key]`.

- `__iter__(self)`  
  Retorna um iterador (`for`, `iter()`).

- `__next__(self)` *(usado em iteradores)*  
  Retorna o próximo valor da iteração.

- `__contains__(self, item)` *(extra útil)*  
  Define o comportamento de `item in obj`.
